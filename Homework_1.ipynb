```python
from ortools.linear_solver import pywraplp

solver = pywraplp.Solver.CreateSolver("GLOP")

# Define the variables.
# w1 = amount of wheat in Feed 1
# w2 = amount of wheat in Feed 2
# a1 = amount of alfalfa in Feed 1
# a2 = amount of alfalfa in Feed 2
w1 = solver.NumVar(0, solver.infinity(), 'w1')
w2 = solver.NumVar(0, solver.infinity(), 'w2')
a1 = solver.NumVar(0, solver.infinity(), 'a1')
a2 = solver.NumVar(0, solver.infinity(), 'a2')

# Set the objective. Explain how you obtained it in the text box after the code block.
# maximizes total profit by summing the revenue generated from selling Feed 1 and Feed 2
# and subtracting the total cost of purchasing the wheat and alfalfa ingredients.
solver.Maximize(
    1.50 * (w1 + a1) + 1.30 * (w2 + a2)
    - 0.50 * (w1 + w2) - 0.40 * (a1 + a2)
)

# Add the constraints.
# Total Wheat Constraint
solver.Add(w1 + w2 <= 1000)
# Total Alfalfa Constraint
solver.Add(a1 + a2 <= 800)
# Feed 1  Constraint
solver.Add(w1 >= 4.0 * a1)
# Feed 2  Constraint
solver.Add(a2 >= 1.5 * w2)
# Solve and print the solution. DO NOT CHANGE ANYTHING PAST THIS POINT.
status = solver.Solve()

if status == pywraplp.Solver.OPTIMAL:
  print("Solution:")
  print(f"Optimal value = {solver.Objective().Value():0.2f}")
  print(f"{w1} = {w1.solution_value():0.2f}")
  print(f"{w2} = {w2.solution_value():0.2f}")
  print(f"{a1} = {a1.solution_value():0.2f}")
  print(f"{a2} = {a2.solution_value():0.2f}")
else:
  print("The problem does not have an optimal solution.")
```

# Problem 5

Read **Winston, p. 120, Problem 46**. Then complete the following code to solve the problem as a linear program.

```python
solver = pywraplp.Solver.CreateSolver("GLOP")

# Complete the definitions of the variables. Use NumVar, not IntVar.
f12 = solver.NumVar(0, solver.infinity(), 'f12') #full time
f1 = solver.NumVar(0, solver.infinity(), 'f1')  # full time
p9 = solver.NumVar(0, solver.infinity(), 'p9') #rest part time
p10 = solver.NumVar(0, solver.infinity(), 'p10')
p11 = solver.NumVar(0, solver.infinity(), 'p11')
p12 = solver.NumVar(0, solver.infinity(), 'p12')
p1 = solver.NumVar(0, solver.infinity(), 'p1')
p2 = solver.NumVar(0, solver.infinity(), 'p2')

# Set the objective
solver.Minimize(64 * (f12 + f1) + 15 * (p9 + p10 + p11 + p12 + p1 + p2))

# Add the constraints
# 9–10
solver.Add(f12 + f1 + p9 >= 4)

# 10–11
solver.Add(f12 + f1 + p9 + p10 >= 3)

# 11–Noon
solver.Add(f12 + f1 + p9 + p10 + p11 >= 4)

# Noon–1 (f12 on lunch)
solver.Add(f1 + p10 + p11 + p12 >= 6)

# 1–2 (f1 on lunch)
solver.Add(f12 + p11 + p12 + p1 >= 5)

# 2–3
solver.Add(f12 + f1 + p12 + p1 + p2 >= 6)

# 3–4
solver.Add(f12 + f1 + p1 + p2 >= 8)

# 4–5
solver.Add(f12 + f1 + p2 >= 8)

# At most five part-time tellers total
solver.Add(p9 + p10 + p11 + p12 + p1 + p2 <= 5)

# Solve and print the solution. DO NOT CHANGE ANYTHING PAST THIS POINT.
status = solver.Solve()

if status == pywraplp.Solver.OPTIMAL:
  print("Solution:")
  print(f"Optimal value = {solver.Objective().Value():0.2f}")
  print(f"{f12} = {f12.solution_value():0.2f}")
  print(f"{f1} = {f1.solution_value():0.2f}")
  print(f"{p9} = {p9.solution_value():0.2f}")
  print(f"{p10} = {p10.solution_value():0.2f}")
  print(f"{p11} = {p11.solution_value():0.2f}")
  print(f"{p12} = {p12.solution_value():0.2f}")
  print(f"{p1} = {p1.solution_value():0.2f}")
  print(f"{p2} = {p2.solution_value():0.2f}")
else:
  print("The problem does not have an optimal solution.")
```
